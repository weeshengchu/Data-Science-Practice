### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='2x', color='#ababab', linewidth=2, marker='.')

plt.title("Our First Graph!", fontdict={'fontname': 'Times New Roman', 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```
